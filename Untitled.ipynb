```python
def depletion(t, I_inf, p, k_er, d):
    return I_inf*(1-p/(2-p)*0.5*np.exp(k_er**2*d**2*0.25)*(np.exp(-k_er*t)*erfc((d**2*k_er-2*t)/(2*d))+np.exp(k_er*t)*erfc((d**2*k_er+2*t)/(2*d))))

def depletion_with_noise(t, I_inf, p, k_er, d):
    I = depletion(t, I_inf, p, k_er, d)
    return I+(np.random.poisson(I, 1)-np.random.poisson(I, 1))/2
    
```

```python
I_inf = 50 # signal amplitude
p = 0.8 # transition probability
d = 0.2 # pulse width in ps
k_er = 1/0.250 # energy transfer rate in ps-1 (typical ET time is ~250fs)

delay = np.linspace(-2, 2, 50)
df = pd.DataFrame(index=["{}ps".format(t) for t in delay], columns=poi_manager_logic.poi_positions.keys())
for name, pos in poi_manager_logic.poi_positions.items():
    pos_dict = {
        'x':pos[0],
        'y':pos[1],
        'z':pos[2],
    }
    scanner_logic.set_target_position(pos_dict)
    #time.sleep(0.1)
    depl = []
    for t in delay:
        motor_logic.move_abs({'x':t+2})
        depl.append(depletion_with_noise(t, I_inf, p, k_er, d)[0])
    df[name] = depl
        
```
